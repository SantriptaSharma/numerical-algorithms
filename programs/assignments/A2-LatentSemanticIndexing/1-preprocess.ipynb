```python
import json
```

```python
books_obj = {}

with open("books_raw.json", "r") as f:
	books_obj = json.load(f)

len(books_obj)
```

Some of the DDNs have trailing or internal whitespace, with some having a pipe character ( | ) at the end

Some books do not have DDNs, or use a custom topic format, eg. 
`"3381": {
    "title": "Vol. I c. 1200 - c.1750  Vol. I",
    "ddn": "Economic history |"
}`, we toss these out to make life easier for me

For similar reasons, we also toss out books with unicode in their titles

```python
import re

cleaned_books = {}

for id, book in books_obj.items():
	ddn = book["ddn"].replace("|", "").replace(" ", "").strip()

	# if empty, discard
	if ddn.isspace():
		print("deleting", book["ddn"], "|", book["title"], "[empty]")
		continue

	# if invalid dewey decimal number, discard
	if re.search("[0-9]{3}(\.[0-9\.]+)?$", ddn) is None:
		print("deleting", book["ddn"], "|", book["title"], "[ddn invalid]")
		continue

	# if any unicode in title, discard
	if not book["title"].isascii():
		print("deleting", book["ddn"], "|", book["title"], "[unicode]")
		continue


	cleaned_books[id] = {
		"title": book["title"],
		"ddn": book["ddn"].replace("|", "").replace(" ", "")
	}

with open("books_cleaned.json", "w") as f:
	json.dump(cleaned_books, f, indent=4)
```
